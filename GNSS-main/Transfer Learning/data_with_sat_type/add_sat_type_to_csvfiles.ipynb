```python
#I would like to use this program to read all simulation data and real world datasets
#And generate new csv files with sat_type column added
#make it easier for SVM model training and evaluation because of fixed sat_type column
import pandas as pd
import numpy as np
import os
import sys
```

```python
current_path = os.getcwd()
print("当前工作目录:", current_path)
parent_dir = os.path.dirname(os.path.dirname(current_path))
print("父目录:", parent_dir)
csv_path = {
        0: os.path.join(parent_dir, "RWD_20250422/D1T4", "observation_data_case5.csv"),
        1: os.path.join(parent_dir, "for_real", "Case1_Urban_10Hz.csv"),
        2: os.path.join(parent_dir, "for_real", "Case1_Suburban_10Hz.csv"),
        3: os.path.join(parent_dir, "for_real", "Case2_Urban_10Hz.csv"),
        4: os.path.join(parent_dir, "for_real", "Case2_Suburban_10Hz.csv"),
        5: os.path.join(parent_dir, "for_real", "Case3_Urban_10Hz.csv"),
        6: os.path.join(parent_dir, "for_real", "Case3_Suburban_10Hz.csv"),
        7: os.path.join(parent_dir, "for_real", "Case4_Urban_10Hz.csv"),
        8: os.path.join(parent_dir, "for_real", "Case4_Suburban_10Hz.csv"),
    }

    
```

```python
output_dir = current_path
os.makedirs(output_dir, exist_ok=True)
print(f" ----------------------------------------------------")
print(f"  - [目标路径] 处理后的文件将保存到: {output_dir}")
print(f" ----------------------------------------------------")
```

```python
for key, path in csv_path.items():
    print(f"  - 处理文件 {path} ...")
    df = pd.read_csv(path)
    df['satellite_id'] = df['satellite_id'].astype(str)
    # 标记 PL (地基)
    is_pl = df['satellite_id'].str.contains("^P")
    # 标记 所有 GNSS (天基)
    is_gps = df['satellite_id'].str.contains("^G", na=False)      #
    is_galileo = df['satellite_id'].str.contains("^E", na=False)  #
    is_beidou = df['satellite_id'].str.contains("^C", na=False)   #
    # 合并 G, E, C 为 "is_gnss"
    is_gnss = is_gps | is_galileo | is_beidou
    # 分配 sat_type (0 = GNSS, 1 = PL, -1 = 其他)
    df['sat_type'] = np.where(
        is_pl, 1,           # 如果是 "P", 设为 1
        np.where(is_gnss, 0, -1) # 否则, 如果是 "G"或"E"或"C", 设为 0
                             # 否则 (比如 "L", "LTE"), 设为 -1
        )
    
    # 4d. 定义输出路径
    base_filename = os.path.basename(path) # (例如 "Case1_Urban_10Hz.csv")
    # 把文件名和扩展名分开
    filename_only, file_extension = os.path.splitext(base_filename)
    # 构造新的文件名
    new_output_filename = f"{filename_only}_with_sat_type{file_extension}"
        
    new_output_path = os.path.join(output_dir, new_output_filename)
    # 4e. 保存
    df.to_csv(new_output_path, index=False, float_format='%.5f')
    print(f" - [成功] Case {key}: 已处理并保存到 {new_output_path}")
    
```
